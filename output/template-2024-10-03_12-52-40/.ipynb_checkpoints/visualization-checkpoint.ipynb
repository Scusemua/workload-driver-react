```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

import os
```

```python
event_durations: np.ndarray[float] = np.loadtxt("./raw_data/session_1/event_durations.txt")
print(event_durations)
```

```python
event_times: np.ndarray[float] = np.loadtxt("./raw_data/session_1/event_times.txt")
print(event_times)
```

```python
inter_arrival_times: np.ndarray[float] = np.loadtxt("./raw_data/session_1/inter_arrival_times.txt")
print(inter_arrival_times)
```

```python
y:int = 1
num_sessions: int = 0
first:bool = True 

ticks: list[int] = []
labels: list[str] = []

linewidth = 15

directory_name:str = "C:/Users/benrc/go/src/github.com/workload-driver-react/output/template-2024-10-03_13-10-30/raw_data"
total_num_sessions: int = len([name for name in os.listdir(directory_name) if os.path.isdir(os.path.join(directory_name, os.fsdecode(name)))])
print("Number of Sessions: %d" % total_num_sessions)

height: int = int(total_num_sessions * 1.5)

plt.style.use('ggplot')
plt.figure(figsize=(height * 2, height), dpi = 256)

directory = os.fsencode(directory_name)
for file in os.listdir(directory):
    filename:str = os.fsdecode(file)
    full_path:str = os.path.join(directory_name, filename)
    print("Found file:", full_path)
    if not os.path.isdir(full_path):
        print(f"\"{full_path}\" is not a directory.")
        continue

    event_durations: np.ndarray[float] = np.loadtxt(os.path.join(full_path, "event_durations.txt"))
    event_times: np.ndarray[float] = np.loadtxt(os.path.join(full_path, "event_times.txt"))
    inter_arrival_times: np.ndarray[float] = np.loadtxt(os.path.join(full_path, "inter_arrival_times.txt"))

    num_sessions += 1
    ticks.append(y)
    labels.append(f"Session {num_sessions}\n{len(event_times)} Events\nAvg IAT: {round(sum(inter_arrival_times)/len(inter_arrival_times), 2)}\nAvg Duration: {round(sum(event_durations)/len(event_durations), 2)}")
    
    for i in range(0, len(event_times)):
        st_tr = event_times[i]
        et_tr = st_tr + event_durations[i]
    
        st_idle = et_tr
    
        try:
            et_idle = et_tr + inter_arrival_times[i]
        except:
            et_idle = et_tr

        if first:
            plt.hlines(y, st_tr, et_tr, color = 'green', linewidth = linewidth, label = "Busy")
            plt.hlines(y, st_idle, et_idle, color = 'grey', linewidth = linewidth, label = "Idle")

            first = False 
        else:
            plt.hlines(y, st_tr, et_tr, color = 'green', linewidth = linewidth)
            plt.hlines(y, st_idle, et_idle, color = 'grey', linewidth = linewidth)

    y += 1

plt.yticks(ticks = ticks, labels = labels)
plt.ylim(0, y)
plt.legend(prop={'size': 24})
```
